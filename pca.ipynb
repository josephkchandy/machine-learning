```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
cancer=load_breast_cancer()
x=cancer.data
y=cancer.target
feature_names=cancer.feature_names
target_names=cancer.target_names
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
pca=PCA(n_components=2)
x_pca=pca.fit_transform(x_scaled)
```

```python
explained_variance=pca.explained_variance_ratio_
print(f"explained variance ratio : {explained_variance}")
```

```python
df_pca=pd.DataFrame(data=x_pca,columns=['PC1','PC2'])
df_pca['target']=y
print("\ndataframe of pca\n")
print(df_pca)
```

```python
plt.figure(figsize=(8,6))
for target,target_name in enumerate(target_names):
 plt.scatter(df_pca[df_pca['target']==target]['PC1'],df_pca[df_pca['target']==target]['PC2'],label=target_name)
plt.xlabel("principal component of x")
plt.ylabel("principal component of y")
plt.title("pca of breast cancer")
plt.legend()
plt.show()
```
